```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.set_printoptions(formatter={'float': lambda x: ('%10.7f')%(x)},
                    suppress=True) #sopprime la notazione scientifica
```

```python
df = pd.read_csv('dati_jacobi.txt',delimiter=' ')
df
```

```python
dim = np.shape(df)
n = dim[0]
```

```python
A = df.iloc[:,:n].values #matrice A dei coefficienti del sistema
b = df.iloc[:,n:].values #vettore b dei termini noti
dim = np.shape(A)
```

```python
def find_max(lista):
    max_value = 0
    pos = 0
    for i in range(len(lista)):
        if lista[i] > max_value:
            max_value = lista[i]
            pos = i
    #ERRORE!!! return lista[i], pos
    return max_value, pos
```

```python
def iter_methods(n = n,eps = 1e-5, printing=True,jacobi = True):
    x2=np.zeros(n)
    max_value = n
    somma = 0
    count = 0
    while(max_value>eps):
        count+=1
        x1 = x2.copy()
        #print(count)
        for j in range(n):
            somma=b[j].copy()
            for k in range(j):
                if jacobi == True: #metodo di Jacobi
                    somma= somma -A[j][k]*x1[k]
                else: #metodo di Gauss-Seidel
                    somma= somma -A[j][k]*x2[k]
            for k in range(j+1,n):
                somma= somma - A[j][k]*x1[k]
            x2[j] = somma/A[j][j]
        diff =abs(x1-x2)
        if printing == True:
            print(("iter. %d, result:")%(count),x2)
        max_value=find_max(diff)[0]
        print("max_value",max_value)
    print(("eps: %s tot. iter. %d, result:")%(eps,count),x2)
    return x2[0]
```

```python
print("2.a) alcuni vettori soluzione (non più di 5, specificando per quale iterazione) \n")
iter_methods()
```
